```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cur_x = 0.1  #Starting point
learning_rate = 0.0001 
precision = 0.000001 #When to stop the algorithm
previous_step_size = 1 
max_iters = 10000 
iters = 0 #iteration counter

#Defining the function f
f = lambda x: x**9 + (3 * (x**8)) - (5 * (x**7)) + (3 * (x**4)) + (7 * (x**2)) - (3 * x)

#Gradient of the function f
df = lambda x: (9 * (x**8)) + (24 * (x**7)) - (35 * (x**6)) + (12 * (x**3)) + (14 * x)

x = np.arange(-10,10,0.00001)
```

```python
plt.title("Plotting the Function")
plt.plot(x,f(x))
plt.grid()
plt.show()
```

```python
plt.title("Plotting the Gradient")
plt.plot(x,df(x))
plt.grid()
plt.show()
```

```python
#Applying Gradient Descent
while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x
    cur_x = cur_x - (learning_rate * df(prev_x))         
    previous_step_size = abs(cur_x - prev_x)
    iters = iters + 1
```

```python
plt.plot(x,f(x))
plt.plot(cur_x,f(cur_x),'r+')
plt.show()

print("The local minima occures at",cur_x)
```
